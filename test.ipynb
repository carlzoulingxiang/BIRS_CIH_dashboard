```python
import numpy as np
import pandas as pd

df = pd.read_excel("/Users/tuanzai/Desktop/Canada/001.UBC/block2.1/BIRS_CIH_dashboard/data/federal-legislation.xlsx")

def handle_err(err, match="{\'vote_text\':"):
    i = len(match)
    n = len(err)
    k = n-i
    
    while True:
        if err[k: k+len(match)] == match:
            break
        k -= 1
    
    return err[:k]+']}'


rst = []
for i, bill_genetic_id in zip(df["votes"], df["goverlytics_id"]):
    if not pd.isna(i):
        try:
            tmp = eval(i)
        except:
            tmp = eval(handle_err(i))
            
    if type(tmp) == tuple:
        tmp = tmp[0]
    vote_values_df = pd.DataFrame(tmp["votes"])
    vote_values_df["bill_genetic_id"] = bill_genetic_id
    for key in ['nv', 'nay', 'yea', 'date', 'total', 'absent', 'passed', 'chamber', 'description']:
        if key in tmp:
            vote_values_df[key] = tmp[key]
        else:
            vote_values_df[key] = np.nan
            
    rst.append(vote_values_df)
    
final_df = pd.concat(rst, axis=0)

final_df.to_csv("./data/federal_bill_vote.csv")   

```
